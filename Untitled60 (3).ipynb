```python
#highest trans amount in butwal
butwal_data = atm_data[atm_data['ATM_Location'] == 'Butwal']
highest_transaction_butwal = butwal_data.loc[butwal_data['Transaction_Amount'].idxmax()]
#Convert the result to a DataFrame for better data
highest_transaction_butwal_df = highest_transaction_butwal.to_frame().T
highest_transaction_butwal_df
```

```python
location_summary = atm_data.groupby('ATM_Location')['Transaction_Amount'].sum().reset_index()

# Create a bar diagram
plt.figure(figsize=(12, 6))
bars = plt.bar(location_summary['ATM_Location'], location_summary['Transaction_Amount'], color='MidnightBlue')

# Add the transaction amount on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.title('Total Transaction Amount by Location')
plt.xlabel('ATM Location')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
atm_data['Transaction_Date'] = pd.to_datetime(atm_data['Transaction_Date'], errors='coerce')

# Extract year and month from Transaction_Date
atm_data['Year'] = atm_data['Transaction_Date'].dt.year
atm_data['Month'] = atm_data['Transaction_Date'].dt.month

# Group by Year and Month, then get the maximum Transaction Amount
highest_transactions_per_month = atm_data.groupby(['Year', 'Month'])['Transaction_Amount'].max().reset_index()

# Rename columns for clarity
highest_transactions_per_month.columns = ['Year', 'Month', 'Highest_Transaction_Amount']
highest_transactions_per_month
```

```python


# Convert Transaction_Amount to numeric (in case it's not)
atm_data['Transaction_Amount'] = pd.to_numeric(atm_data['Transaction_Amount'], errors='coerce')

# 1. Highest and lowest transaction amounts by location
highest_by_location = atm_data.groupby('ATM_Location')['Transaction_Amount'].max().reset_index()
lowest_by_location = atm_data.groupby('ATM_Location')['Transaction_Amount'].min().reset_index()

# Renaming columns
highest_by_location.columns = ['ATM_Location', 'Highest_Transaction_Amount']
lowest_by_location.columns = ['ATM_Location', 'Lowest_Transaction_Amount']

# Merging highest and lowest
transaction_summary = pd.merge(highest_by_location, lowest_by_location, on='ATM_Location')
transaction_summary
```

```python
 # 2. Plotting a pie chart for highest transaction amounts
plt.figure(figsize=(14, 6))
plt.pie(transaction_summary['Highest_Transaction_Amount'], labels=transaction_summary['ATM_Location'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Highest Transaction Amounts by ATM Location')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(transaction_summary['ATM_Location'], title='ATM Locations', loc='upper right')
plt.show()

# 3. Plotting a pie chart for lowest transaction amounts
plt.figure(figsize=(14, 6))
plt.pie(transaction_summary['Lowest_Transaction_Amount'], labels=transaction_summary['ATM_Location'], autopct='%1.1f%%', startangle=230)
plt.title('Distribution of Lowest Transaction Amounts by ATM Location')
plt.axis('equal')
plt.legend(transaction_summary['ATM_Location'], title='ATM Locations', loc='upper right')
plt.show()
```

```python

# Count the number of successful, failed, and cancelled transactions
transaction_counts = atm_data['Transaction_Status'].value_counts()

# Extract counts
successful_count = transaction_counts.get('Successful', 0)
failed_count = transaction_counts.get('Failed', 0)
cancelled_count = transaction_counts.get('Cancelled', 0)

# Calculate total transactions
total_transactions = successful_count + failed_count + cancelled_count

# Calculate success rate
if total_transactions > 0:
    success_rate = (successful_count / total_transactions) * 100
else:
    success_rate = 0

# Create a summary DataFrame
summary_data = {
    'Transaction Type': [
        'Total Transactions',
        'Successful Transactions',
        'Failed Transactions',
        'Cancelled Transactions',
        'Success Rate (%)'
    ],
    'Count': [
        total_transactions,
        successful_count,
        failed_count,
        cancelled_count,
        success_rate
    ]
}

summary_df = pd.DataFrame(summary_data)
summary_df
```

```python

# Count the number of successful, failed, and cancelled transactions
transaction_counts = atm_data['Transaction_Status'].value_counts()

# Extract counts
successful_count = transaction_counts.get('Successful', 0)
failed_count = transaction_counts.get('Failed', 0)
cancelled_count = transaction_counts.get('Cancelled', 0)

# Calculate total transactions
total_transactions = successful_count + failed_count + cancelled_count

# Calculate success rate
if total_transactions > 0:
    success_rate = (successful_count / total_transactions) * 100
else:
    success_rate = 0

# Create a summary DataFrame
summary_data = {
    'Transaction Type': [
        'Total Transactions',
        'Successful Transactions',
        'Failed Transactions',
        'Cancelled Transactions',
        
        'Success Rate (%)'
    ],
    'Count': [
        total_transactions,
        successful_count,
        failed_count,
        cancelled_count,
        success_rate
    ]
}

summary_df = pd.DataFrame(summary_data)
summary_df
```

```python

# Plotting the bar diagram
plt.figure(figsize=(10, 5))
bars = plt.bar(summary_df['Transaction Type'], summary_df['Count'], color=['green', 'red', 'orange'])

plt.title('ATM Transactions by Status')
plt.xlabel('Transaction Status')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Adding the count on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.show()
```

```python
# Calculate total transactions, successful transactions, failed transactions for each bank
bank_performance = atm_data.groupby('Bank_Name').agg(
    total_transactions=('Transaction_Status', 'count'),
    successful_transactions=('Transaction_Status', lambda x: (x == 'Successful').sum()),
    failed_transactions=('Transaction_Status', lambda x: (x == 'Failed').sum()),
    average_transaction_amount=('Transaction_Amount', 'mean')
).reset_index()

# Calculate success rate
bank_performance['success_rate (%)'] = (bank_performance['successful_transactions'] / bank_performance['total_transactions']) * 100
```

```python
 bank_performance
```

```python
plt.figure(figsize=(14, 9))

# Bar plot for total, successful, and failed transactions
plt.subplot(2, 1, 1)
bar_width = 0.3
index = range(len(bank_performance))

plt.bar(index, bank_performance['total_transactions'], bar_width, label='Total Transactions', color='blue')
plt.bar([i + bar_width for i in index], bank_performance['successful_transactions'], bar_width, label='Successful Transactions', color='green')
plt.bar([i + 2 * bar_width for i in index], bank_performance['failed_transactions'], bar_width, label='Failed Transactions', color='red')

plt.xlabel('Banks')
plt.ylabel('Number of Transactions')
plt.title('Transactions by Bank')
plt.xticks([i + bar_width for i in index], bank_performance['Bank_Name'], rotation=45)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
 transaction_type_summary = atm_data.groupby('Transaction_Type').agg(
total_amount=('Transaction_Amount', 'sum'),
total_transactions=('Transaction_ID', 'count')
).reset_index()

# Visualize total transaction amounts by Transaction Type
plt.figure(figsize=(12, 6))
sns.barplot(x='Transaction_Type', y='total_amount', data=transaction_type_summary, palette='coolwarm')
plt.title('Total Transaction Amounts by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Ensure Transaction_Time is in datetime format
atm_data['Transaction_Time'] = pd.to_datetime(atm_data['Transaction_Time'], format='%H:%M:%S')

# Extract hour from Transaction_Time
atm_data['Hour'] = atm_data['Transaction_Time'].dt.hour

# Group by hour and count transactions
transaction_counts_by_hour = atm_data.groupby('Hour').size().reset_index(name='Transaction_Count')

# Sort by hour for better visualization
transaction_counts_by_hour = transaction_counts_by_hour.sort_values(by='Hour')
transaction_counts_by_hour
```
